# Part 1 — Brightness/contrast adjustment of high-speed frames

Applies OpenCV `convertScaleAbs` (contrast factor 1.2, brightness offset 60) to the high-speed camera frames so the shuttlecock–racket contact is easier to read. Point the input path at your own frame folder before running.

```python
import cv2
import os
from pathlib import Path

ALPHA = 1.2  # contrast
BETA = 60    # brightness

base_dir = Path(__file__).parent
output_root = base_dir / 'changed_image'
output_root.mkdir(exist_ok=True)

for i in range(1, 8):
    input_folder = base_dir / f'test{i}' / 'image'
    output_folder = output_root / f'test{i}' / 'image'
    output_folder.mkdir(parents=True, exist_ok=True)
    for img_file in input_folder.glob("*"):
        if img_file.suffix.lower() not in ['.jpg', '.jpeg', '.png']:
            continue 
        image = cv2.imread(str(img_file))
        if image is None:
            print(f"Error:{img_file}")
            continue
        adjusted = cv2.convertScaleAbs(image, alpha=ALPHA, beta=BETA)
        output_path = output_folder / img_file.name
        cv2.imwrite(str(output_path), adjusted)
```
